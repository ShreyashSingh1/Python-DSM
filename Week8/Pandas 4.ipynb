```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
}
)

data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
data1
```

```python
data2
```

```python
# Inner merge
pd.merge(data1, data2)
```

```python
pd.merge(data1, data2, how="left", on="key1")
```

```python
pd.merge(data1, data2, how="right")
```

```python
pd.merge(data1, data2, how="outer")
```

```python
# Cartesian product leta hai
pd.merge(data1, data2, how="cross")
```

```python
data1 = pd.DataFrame({
    'key1':[1,2,4,5,6],
    'key2':[4,5,6,7,8],
    'key3':[3,4,5,6,6]
},index = ['a','b','c','d','e'])

data2 = pd.DataFrame({
    'key11':[1,2,4,5,6],
    'key22':[4,5,6,7,8],
    'key33':[3,4,5,6,6]
},index=['a','b','h','i','j'])
```

```python
# Default left join, Basically it works on basis of indexes and merge works on bases of column
data1.join(data2)
```

```python
data1.join(data2, how="right")
```

```python
data1.join(data2, how="outer")
```

```python
# Cartesian Product nhi leta bas mapping kar dena hai
data1.join(data2, how="cross")
```
