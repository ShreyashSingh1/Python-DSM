## Time Resampling 

### Time resampling is like adjusting a clock to tick at a different speed. In data terms, it means changing how often you look at your information.
Time resampling refers to the process of changing the frequency of a time series. In the context of time series data, observations are recorded at specific time intervals. Resampling allows you to adjust the frequency of these observations, either by aggregating data over larger intervals (downsampling) or by interpolating data to create observations at smaller intervals (upsampling).

Original Data (Daily):

Day 1: $50
Day 2: $55
Day 3: $52
Day 4: $58
Day 5: $60
Now, let's explore both downsampling and upsampling:

Downsampling (Aggregation - Weekly):

If we want to see the average stock price on a weekly basis:
Week 1: ($50 + $55 + $52) / 3 = $52.33
Week 2: ($58 + $60) / 2 = $59
We've reduced the number of data points from daily to weekly, summarizing the information.

Upsampling (Interpolation - Hourly):

If we want to estimate the stock prices on an hourly basis:
Between Day 1 and Day 2: Linearly interpolate values (assuming a steady increase): $50, $52.50, $55
Between Day 2 and Day 3: $55, $53.50, $52
And so on...
We've increased the frequency of data points by estimating values between the original daily data points.
In these examples, downsampling and upsampling help adjust the time intervals to better suit the analysis or visualization needs. It's like zooming in or out on the timeline of your data.

Time resampling is a common preprocessing step in time series analysis to align data with different time frequencies or to create a more uniform dataset for modeling purposes. The choice of resampling method depends on the specific characteristics of the data and the goals of the analysis.

https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```python
tsla_data.resample(rule='A').min()
```

```python
tsla_data.resample(rule='A').max()
```

```python
# Rule A is for Annual start frequency
tsla_data.resample(rule='A').max()["Open"].plot()
```

```python
# Rule QS is for quarterly start frequency
tsla_data.resample(rule='QS').max()["Open"].plot()
```

```python
tsla_data.resample(rule='BA').max()
```

```python
# Rule BA is for yearly End frequency
tsla_data.resample(rule='BA').max()["High"].plot()
```

```python
tsla_data["High"].resample(rule='BA').mean().plot(kind='bar')
```

```python
tsla_data["High"].resample(rule='M').mean().plot(kind='bar', figsize=(14, 4), color='lightblue')
```

```python
# It is like a moving average
tsla_data["High"].rolling(window=10).mean().head(20)
```

```python
tsla_data["Open 30 Days Rolling"] = tsla_data["Open"].rolling(window=30).mean()
tsla_data.head(40)
```

```python
tsla_data[["Open", "Open 30 Days Rolling"]].plot(figsize=(14, 4))
```
