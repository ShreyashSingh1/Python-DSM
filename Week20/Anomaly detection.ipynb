### Anomaly detection

```python
import pandas as pd
df = pd.read_csv('C:/Users/shrey/OneDrive/Documents/Skills/Week20/Dataset/healthcare.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(df.iloc[:,0], df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
clf = IsolationForest(contamination=0.2)
clf.fit(df)
predictions = clf.predict(df)
predictions
```

-1 outliers,
1 normal data points

```python
import numpy as np
index = np.where(predictions < 0)
index
```

```python
x = df.values 
x
```

```python
index = np.where(predictions < 0)
plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.scatter(x[index,0], x[index,1], edgecolors="r")
```

```python
sns.scatterplot(x=df.iloc[:,0], y=df.iloc[:,1], hue=predictions)
```
