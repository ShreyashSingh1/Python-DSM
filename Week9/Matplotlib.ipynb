```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y, '--b')
plt.xlabel("This x data")
plt.ylabel("This y data")
plt.grid()
plt.title("This is my title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.rand(50)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Random data")
# Alpha is intensity of dots
plt.scatter(x, y, c="b", s=sizes*200, alpha=.5, marker="v")
```
