# Handling Outliers
### 5 Number Summary
1. Minimum Value
2. Q1- 25 percentile
3. Median
4. Q3- 75 percentile
5. Maximum Value



```python
import numpy as np
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]
```

```python
"""Array aur fir jo bhi %tile chaiye vo pass karne ka"""
a = np.percentile(lst_marks, [25])
b = np.percentile(lst_marks, [75])
c = np.percentile(lst_marks, [100])
d = np.percentile(lst_marks, [50])
e = np.percentile(lst_marks, [0])

a, b, c, d, e
```

```python
## [Lower Fence <---> Higher Fence] 
Q1 = np.percentile(lst_marks, [25])
Q1
```

```python
min, q1, q2, q3, max = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
max
```

```python
# So that we can calculate lower/upper fence
IQR = q3 - q1
IQR
```

```python
lower_fence = q1 - 1.5*IQR
higher_fence = q3 + 1.5*IQR

lower_fence, higher_fence
```
